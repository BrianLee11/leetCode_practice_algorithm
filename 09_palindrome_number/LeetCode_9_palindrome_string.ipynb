Given an integer x, return true if x is a palindrome, and false otherwise.

Example 1:

Input: x = 121

Output: true

Explanation: 121 reads as 121 from left to right and from right to left.


Example 2:

Input: x = -121

Output: false

Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.


Example 3:

Input: x = 10

Output: false

Explanation: Reads 01 from right to left. Therefore it is not a palindrome.

# Answer

# Submission to LeetCode

```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """

        # case 1) check negativity
        if x < 0:
          return False

        # case 2) if non-negative number
        else:

          # convert x to a string
          x = str(x)

          # compare the leftmost character and rightmost character
          # a//b gives a quotient part.
          for index in range(len(x)//2):

            # checking non-equal characters.
            # If unequal, this number is not a palindrome number
            if x[index] != x[len(x)-1]:
              return False

        # Otherwise, palindrome number
        return True
```

```python
# testing with example
# example1: x = 121. x is a palindrome number.
# result should be True

# initialize class Solution
solution = Solution()

print(solution.isPalindrome(121))
```

```python
# testing with example
# example2: x = -121. x is a negative number, therefore not a palindrome number
# result should be False

# initialize class Solution
solution = Solution()

print(solution.isPalindrome(-121))
```

```python
# testing with example
# example3: x = 10. x is non-mirrored number, therefore not a palindrome number
# result should be False

# initialize class Solution
solution = Solution()

print(solution.isPalindrome(10))
```

```python
x = -121

x = str(x)

for index in range(len(x)//2 +1):

  # a negative number
  if x[0] == '-':
    print("Not a palindrome number")
    break

  # a normal case: check whether the left-most character equals to the right-most character
  if x[index] != x[len(x)-1-index]:
    print("Not a palindrome number")
    break

  # reaches the middle character
  else:
    if index == len(x) //2:
      print("A palindrome number")
```

# Coding with function

```python
def isPalindrome(x):
    """
    :type x: int
    :rtype: bool
    """

    # check for the negative number
    if x < 0:
      return False

    # non-negative number x
    # transform integer x to a string x
    x = str(x)

    for index in range(len(x)//2 +1):
      if x[index] != x[len(x)-1-index]:
        return False

    return True
```

```python
# testing
isPalindrome(121)
```

```python
# testing
isPalindrome(-121)
```

```python
# testing
isPalindrome(10)
```
