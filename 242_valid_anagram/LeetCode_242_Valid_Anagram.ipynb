Given two strings s and t, return true if t is an anagram of s, and false otherwise.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Example 1:**

- Input: s = "anagram", t = "nagaram"
- Output: true

**Example 2:**

- Input: s = "rat", t = "car"
- Output: false

# Codes submitted to LeetCode

````python
class Solution(object):
    def isAnagram(self, string1, string2):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        dict1 = {}

        for index in range(len(string1)):
            if string1[index] in dict1:
                dict1[string1[index]] += 1
            else:
                dict1[string1[index]] = 1

        dict2 = {}

        for index in range(len(string2)):
            if string2[index] in dict2:
                dict2[string2[index]] += 1
            else:
                dict2[string2[index]] = 1

        return dict1 == dict2
````

````python
# testing
solution = Solution()
result = solution.isAnagram("anagram", "nagaram")
print(result)

result = solution.isAnagram("rat", "car")
print(result)
````

# Analysis: time complexity

**1.Initialization of dictionary: $O(1)$**

`dict1 = {}`

**2. the first for loop is $O(n)$ where n is the length of string1**

- `len(string1)` is $O(1)$ because for given array string, len(string) calls an attribute of array string. Therefore, len() is $O(1)$.

- `for index1 in range()` is $O(len(string)) = O(n)$.

```
            if string1[index1] in dict:    # O(1)
                dict[string[index1]] += 1  # O(1)
            else:
                dict[string[index1]] = 1   # O(1)
```

**3. similarly, the second for loop is $O(m)$ where m is the length of string2**

**4. the return part is $O(1)$**            
        `return dict1 == dict2`

**The overall time complexity is:** $$O(n) + O(m) + O(1) = O(n+m)$$
