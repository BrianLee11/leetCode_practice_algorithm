````python
class Solution(object):
    def productExceptSelf(self, array):
        """
        :type nums: List[int]
        :rtype: List[int]
        time complexity: O(n), where n is the length of array
        intuition: for ith element in the given array, compute the product of numbers from the beginning to i-1 and name this product as prefix.
        Compute the product of numbers from the end to i+1 and name this product as suffix.
        Now, the product of numbers except the ith element would be the product of prefix and suffix.

        Example: array = [1,2,3,4,5,6]
        Let i = 2 (array[2] = 3)
          prefix = array[0] * ... * array[i-1] = 1 * 2 = 2
          suffix = array[length of array - 1] * array[length of array - 2] * ... * array[i+1] = 6 * 5 * 4 = 120
        product of numbers except self = prefix * suffix = 2 * 120 = 240 = 1 * 2 * 4 * 5 * 6
        """
        # compute the length of array
        lengthArray = len(array)         # O(n), where n is length of array

        # initialize prefix and suffix arrays
        prefix = [1] * lengthArray       # O(n), where n is length of array
        suffix = [1] * lengthArray       # O(n), where n is length of array

        # construct prefix array
        # Time complexity of for loop below: O(n), where n is length of array
        for index in range(lengthArray): # O(n), where n is length of array
            if index == 0:               # O(1)
                prefix[0] = 1            # O(1)
            elif index == 1:             # O(1)
                prefix[1] = array[0]     # O(1)
            else:
                prefix[index] = prefix[index-1] * array[index-1] # O(1)

        # construct suffix array
        # Time complexity of for loop below: O(n), where n is length of array
        for index in range(lengthArray - 1, -1, -1):  # O(n), where n is length of array
            if index == lengthArray - 1:              # O(1)
                suffix[index] = 1                     # O(1)
            elif index == lengthArray - 2:            # O(1)
                suffix[index] = array[lengthArray-1]  # O(1)
            else:
                suffix[index] = suffix[index+1] * array[index+1]

        # initialize product array
        productArray = [1] * lengthArray              # O(n)

        # generate productArray
        # Time complexity of for loop below: O(n), where n is length of array
        for index in range(lengthArray):                         # O(n)
            productArray[index] = prefix[index] * suffix[index]  # O(1)

        # return the products in array
        return productArray
````

````python
# Testing with example 1
nums = [1,2,3,4]
solution = Solution()

result = solution.productExceptSelf(nums) # output should be [24,12,8,6]
print(result)
````

````python
# Testing with example 2
nums = [-1,1,0,-3,3]
solution = Solution()

result = solution.productExceptSelf(nums) # output should be [0,0,9,0,0]
print(result)
````

# **Intuition**

- For ith element in the given array, compute the product of numbers from the beginning to i-1 and name this product as **prefix**.
        
- Compute the product of numbers from the end to i+1 and name this product as **suffix**.

- Now, the product of numbers except the ith element would be the **product of prefix and suffix**.

        Example: array = [1,2,3,4,5,6]
        Let i = 2 (array[2] = 3)
          prefix = array[0] * ... * array[i-1] = 1 * 2 = 2
          suffix = array[length of array - 1] * array[length of array - 2] * ... * array[i+1] = 6 * 5 * 4 = 120
        product of numbers except self = prefix * suffix = 2 * 120 = 240 = 1 * 2 * 4 * 5 * 6

#**Analysis: time complexity**

**1. compute the length of array and initialization**
```
        # compute the length of array
        lengthArray = len(array)         # O(n), where n is length of array

        # initialize prefix and suffix arrays
        prefix = [1] * lengthArray       # O(n), where n is length of array
        suffix = [1] * lengthArray       # O(n), where n is length of array
```

- Counting the number of elements in the array is $O(n)$, where n is the length of array.

- Initialization of prefix and suffix arrays is also $O(n)$

**2. Construction of the prefix array**
```
        # construct prefix array
        # Time complexity of for loop below: O(n), where n is length of array
        for index in range(lengthArray): # O(n), where n is length of array
            if index == 0:               # O(1)
                prefix[0] = 1            # O(1)
            elif index == 1:             # O(1)
                prefix[1] = array[0]     # O(1)
            else:
                prefix[index] = prefix[index-1] * array[index-1] # O(1)
```

- Since each element in the prefix array is visited, the outer time complexity of $O(n)$.

- Inside the loop, the most complex operation is $O(1)$ because all operations inside the loop are constant-time operations.

- The overall time complexity is $O(n * 1) = O(n)$.



**3. Construction of the suffix array**
```
        # construct suffix array
        # Time complexity of for loop below: O(n), where n is length of array
        for index in range(lengthArray - 1, -1, -1):  # O(n), where n is length of array
            if index == lengthArray - 1:              # O(1)
                suffix[index] = 1                     # O(1)
            elif index == lengthArray - 2:            # O(1)
                suffix[index] = array[lengthArray-1]  # O(1)
            else:
                suffix[index] = suffix[index+1] * array[index+1]
```
- Similar to the previous part 2. construction of the prefix array, the time complexity of this part 3 is also $O(n)$.


**4. The rest of codes: initialization and generating product is also O(n)**
```
        # initialize product array
        productArray = [1] * lengthArray              # O(n)

        # generate productArray
        # Time complexity of for loop below: O(n), where n is length of array
        for index in range(lengthArray):                         # O(n)
            productArray[index] = prefix[index] * suffix[index]  # O(1)
```

- Initialization n-number of array is $O(n)$.

- Multiplying prefix and suffix n-times is also $O(n)$.

        # return the products in array
        return productArray

# Overall time complexity is $O(n)$
- where n is the number of elements in the array.
- The dominant (most complex) time complexity is $O(n)$ in the codes. Therefore the time complexity is $O(n)$.
