14. Longest common prefix
- Write a function to find the longest common prefix string amongst an array of strings.
- If there is no common prefix, return an empty string ""

**Example 1**
- Inputs: strs = ["flower", "flow", "flight"]
- Output:"fl"

**Example 2**
- Inputs: strs = ["dog", "racecar", "car"]
- Outputs: ""
- Explanation: There is no common prefix among the input strings.

#**LeetCode submission**
# Codes:



```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :input: list of strings
        :output: string with largest common prefix characters
        :motivation:
          1. sort the list,
          2. compare the first and last element in the list,
          3. count the number of characters that share the same characters between the first and last element
          4. return the subset of first element's characters
        :time complexity:O(n log(n)) where n:= the number of elements in the list
        """

        # sort the strings in ascending order
        strs.sort()

        # set the first and last strings
        firstWord = strs[0]
        lastWord = strs[-1]

        # keep the index
        index = 0

        # iterate until a different character appears
        while index < len(firstWord) and index < len(lastWord) and firstWord[index] == lastWord[index]:
            index = index + 1

        # the longest common prefix
        longestCommonPrefix = firstWord[:index]

        return longestCommonPrefix
```

```python
#Example 1
#Inputs: strs = ["flower", "flow", "flight"]
#Output:"fl"

strs = ["flower", "flow", "flight"]

solution = Solution()
output = solution.longestCommonPrefix(strs)
print(f'Input: {strs}, Output: {output}')

#Example 2
#Inputs: strs = ["dog", "racecar", "car"]
#Outputs: ""
#Explanation: There is no common prefix among the input strings.

strs = ["dog", "racecar", "car"]
output = solution.longestCommonPrefix(strs)
print(f'Input: {strs}, Output: {output}')
```

```python
# start with a simple case of two words
listWords = ["flower", "flow"]

# sort the strings in the list of words.
# sorting would facilitate the largestCommonPrefix operation
listWords.sort()

# Compare each character of two words, one-by-one
index = 0

# calculate the index where the furthest common character is reached
while index < len(listWords[0]) and index < len(listWords[1]) and listWords[0][index] == listWords[1][index]:
  index = index + 1

# Compute the largest common prefix
largestCommonPrefix = listWords[0][:index]

print(largestCommonPrefix) #output: flow
```

```python
# move with a case of three words, with similar logic in the previous case
listWords = ["flower", "flow", "flight"]

# sort the strings
listWords.sort()

index = 0

while index < len(listWords[0]) and index < len(listWords[1]) and index < len(listWords[2]) and listWords[0][index] == listWords[1][index] == listWords[2][index]:
  index = index + 1

largestCommonPrefix = listWords[0][:index]

print(largestCommonPrefix) #output: fl
```
