#Question 347: top k-frequent elements

Given an integer array nums and an integer k, return the k most frequent elements. You may return the answer in any order.

**Example 1:**
- Input: nums = [1,1,1,2,2,3], k = 2
- Output: [1,2]

**Example 2:**
- Input: nums = [1], k = 1
- Output: [1]

**Follow up**: Your algorithm's time complexity must be better than $O(n \cdot log n)$, where n is the array's size.

````python
# https://leetcode.com/problems/top-k-frequent-elements/description/
# 347. Top K Frequent Elements

class Solution(object):
    def topKFrequent(self, nums, k):
        # Initialize a dictionary to store the frequency of each number in nums
        dictionary = {}

        # Iterate through nums and count the occurrences of each number
        for number in nums:
            if number in dictionary:
                dictionary[number] += 1
            else:
                dictionary[number] = 1

        # Convert the frequency values and keys of the dictionary to lists
        frequencyValues = list(dictionary.values())
        keys = list(dictionary.keys())

        # Initialize attempt counter and list to store the top k frequent elements
        attempt = 0
        maxElements = []

        # Loop until we find the top k elements
        while attempt < k:
            # Find the maximum frequency value
            maxFrequency = max(frequencyValues)
            # Find the index of the maximum frequency value
            maxIndex = frequencyValues.index(maxFrequency)
            # Append the corresponding key (number) to the maxElements list
            maxElements.append(keys[maxIndex])

            # Set the frequency of the found maximum element to -1 to exclude it from future iterations
            frequencyValues[maxIndex] = -1
            # Increment the attempt counter
            attempt += 1

        # Return the list of top k frequent elements
        return maxElements
````

````python
# Example usage
solution = Solution()
nums = [4, 3, 4, 4, 4, 4, 1, 1, 2, 2, 3]
k = 2
print(solution.topKFrequent(nums, k))  # Output: [4, 3]
````
