Given an array of strings strs, group the anagrams together. You can return the answer in any order.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

**Example 1:**

- Input: strs = ["eat","tea","tan","ate","nat","bat"]
- Output: [["bat"],["nat","tan"],["ate","eat","tea"]]

**Example 2:**

- Input: strs = [""]
- Output: [[""]]

**Example 3:**

- Input: strs = ["a"]
- Output: [["a"]]

# **Codes submitted to LeetCode**

```python
class Solution(object):
    def groupAnagrams(self, array):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """

        dictionary = {}

        for string in array:
            sorted_string = sorted(string)
            sorted_string = ''.join(sorted_string)

            if sorted_string in dictionary:
                dictionary[sorted_string].append(string)
            else:
                dictionary[sorted_string] = [string]

        return dictionary.values()
```

```python
array = ["eat","tea","tan","ate","nat","bat"]

solution = Solution()
result = solution.groupAnagrams(array)
print(result)
```

#**Analysis: time complexity**

```python
        for string in array:                               # O(n), where n is number of elements in array
            sorted_string = sorted(string)                 # k log(k), where k is the maximum number of characters
            sorted_string = ''.join(sorted_string)         # O(1) where concantenation is O(1)

            if sorted_string in dictionary:                # if ... then statement is O(1)
                dictionary[sorted_string].append(string)   # add value in key-value is O(1)
            else:
                dictionary[sorted_string] = [string]       # insert value in key-value is O(1)
```

**1.initlization of dictionary**

- `dictionary = {}`

This initialization is $O(1)$.

**2. the for loop**

- `for string in array:`

This is $O(n)$, where $n$ is the number of elements in the array.

**3. Inside the for loop: sorted(string)**

- `sorted(string)` is $O(k \cdot \log(k))$ where k is the number of maximum elements in the strings. (related: see quicksort algorithm or mergesort algorithm)

**4. Inside the for loop: join(sorted_string)**
- This is $O(1)$.

**5. Inside the for loop: if ... then statement**
- `if sorted_string in dictionary:` is $O(1)$.
- `dictionary[sorted_string].append(string)' is $O(1)$
- `else dictionary[sorted_string] = [string]` is $O(1)$

**6. return dictionary.values()** is $O(1)$

#**Overall time complexity:**

$$n \cdot k \cdot \log(k)$$ where n = number of elements in the array, k = the maximum number of characters in the elements in the array.
