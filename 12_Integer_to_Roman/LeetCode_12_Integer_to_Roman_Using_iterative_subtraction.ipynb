Seven different symbols represent Roman numerals with the following values:

Symbol	Value
- I	1
- V	5
- X	10
- L	50
- C	100
- D	500
- M	1000

Roman numerals are formed by appending the conversions of decimal place values from highest to lowest. Converting a decimal place value into a Roman numeral has the following rules:

- If the value does not start with 4 or 9, select the symbol of the maximal value that can be subtracted from the input, append that symbol to the result, subtract its value, and convert the remainder to a Roman numeral.

- If the value starts with 4 or 9 use the subtractive form representing one symbol subtracted from the following symbol, for example, 4 is 1 (I) less than 5 (V): IV and 9 is 1 (I) less than 10 (X): IX. Only the following subtractive forms are used: 4 (IV), 9 (IX), 40 (XL), 90 (XC), 400 (CD) and 900 (CM).

- Only powers of 10 (I, X, C, M) can be appended consecutively at most 3 times to represent multiples of 10. You cannot append 5 (V), 50 (L), or 500 (D) multiple times. If you need to append a symbol 4 times use the subtractive form.

- Task: Given an integer, convert it to a Roman numeral.

#Codes submitted to LeetCode

````python
# LeetCode question 12: "Integer to Roman"
# url: https://leetcode.com/problems/integer-to-roman/description/
class Solution(object):
    def intToRoman(self, integer):
        """
        :type num: int. 1 to 3999
        :rtype: strs
        """

        # Create a dictionary to store corresponding values of integers to Roman numerals.
        # Time complexity: O(1), since this dictionary contains a fixed set of key-value pairs
        dictionary_digit_Roman = {
            1000: 'M', 900: 'CM', 500: 'D', 400: 'CD',
            100: 'C', 90: 'XC', 50: 'L', 40: 'XL',
            10: 'X', 9: 'IX', 5: 'V', 4: 'IV', 1: 'I'
        }

        # Initialize an array to store the resulting Roman numeral characters.
        # This initialization is also O(1)
        array = []

        # Iterate over the dictionary keys sorted in descending order.
        """
        Sorting n elements takes O(n log(n)) times
            When n is fixed and small (like 13), the sorting time is O(13 log(13)), which is a constant
            O(13 log (13)) is a constant, therefore O(1)

        for key ... loop takes O(13), which is O(1)
        """
        for key in sorted(dictionary_digit_Roman.keys(), reverse=True):

            # While the current integer value is greater than or equal to the key,
            # append the corresponding Roman numeral to the array and subtract the key from the integer.
            """
            While this loop’s iterations depend on the value of integer,
            the maximum value for integer in the context of Roman numeral
            conversion is bounded (e.g., typically 1 to 3999).
            """
            while integer >= key:
                array.append(dictionary_digit_Roman[key])
                integer = integer - key

        # Join the array into a single string and return it.
        return ''.join(array) # This is O(1)
````

# Testing

````python
solution = Solution()
result = solution.intToRoman(3749)
print(result)
print(result == "MMMDCCXLIX")

result = solution.intToRoman(58)
print(result)
print(result == "LVIII")

result = solution.intToRoman(1994)
print(result)
print(result == "MCMXCIV")
````

##**Intuition Behind the Code**

**1. Understanding Roman Numerals:**

- Roman numerals are represented by combinations of letters from the Latin alphabet: I (1), V (5), X (10), L (50), C (100), D (500), M (1000).

- Some numerals are exceptions, like IV (4), IX (9), XL (40), XC (90), CD (400), CM (900).

**2. Building the Roman Numeral:**

- The function iterates over the dictionary keys, which are sorted in descending order. For each key, it repeatedly subtracts the key from the value and appends the corresponding Roman numeral to the array until the value is less than the key.

- This ensures that the largest possible Roman numeral is used first, followed by smaller numerals.

#Step by step explanation

**1. Initialize Variables and Dictionaries:**

````python
dictionary_digit_Roman = {1000: 'M', 900: 'CM', 500: 'D', 400: 'CD', 100: 'C', 90: 'XC', 50: 'L', 40: 'XL', 10: 'X', 9: 'IX', 5: 'V', 4: 'IV', 1: 'I'}
array = []
value = integer
````

**2. Iterate Over the Dictionary and Build the Roman Numeral:**

````python
for key in dictionary_digit_Roman:
    while value >= key:
        array.append(dictionary_digit_Roman[key])
        value = value - key
````

**3.Join the Resulting Array into a String and Return:**

````python
return ''.join(array)
````


**2. Sorting the Keys**

`for key in sorted(dictionary_digit_Roman.keys(), reverse=True):`

Sorting a fixed number of elements is a constant time operation because the number of elements is constant. In computational complexity terms:

- Sorting  n elements generally takes O(nlogn) time.
- When n is fixed and small (like 13), the sorting time is O(13log13), which is a constant.

Therefore, sorting these 13 keys is considered an O(1) operation because the time it takes does not grow with the size of the input; it remains constant.

**3. Iterating Over the Keys**
Similarly, iterating over the sorted keys:

```
for key in sorted(dictionary_digit_Roman.keys(), reverse=True):
```
This loop runs 13 times, once for each key. Since the number of iterations is fixed and does not depend on the input size, the time complexity is O(1).

**4.Combined Effect**
- Sorting 13 keys: O(1) because the number of keys is fixed.
- Iterating over 13 keys: O(1) because the number of iterations is fixed.

The inner while loop executes a number of times proportional to the value of integer:

```
while integer >= key:

    integer = integer - key
```

While this loop’s iterations depend on the value of integer, the maximum value for integer in the context of Roman numeral conversion is bounded (e.g., typically 1 to 3999).
