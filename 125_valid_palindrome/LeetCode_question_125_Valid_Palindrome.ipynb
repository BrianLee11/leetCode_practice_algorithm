A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers.

Given a string s, return true if it is a palindrome, or false otherwise.

**Example 1:**
- Input: s = "A man, a plan, a canal: Panama"
- Output: true
- Explanation: "amanaplanacanalpanama" is a palindrome.

**Example 2:**
- Input: s = "race a car"
- Output: false
- Explanation: "raceacar" is not a palindrome.

**Example 3:**
- Input: s = " "
- Output: true
- Explanation: s is an empty string "" after removing non-alphanumeric characters.

Since an empty string reads the same forward and backward, it is a palindrome.

**Constraints:**
- 1 <= s.length <= $2 * 10^5$
- s consists only of printable ASCII characters.

#**Codes submitted to LeetCode**

````python
class Solution(object):
    def isPalindrome(self, string):
        """
        :input type string1: str
        :output type: bool
        :Total time complexity: O(n), where
        """
        import re

        # Consider lowercase only
        # time complexity: O(n) where n is the number of characters in the string
        # each character is visisted. If uppercase alphabet character, make it a lowercase: O(1)
        # If not uppercase(thus lowercase), leave as it is: O(1)
        lowercase_string = string.lower()

        # Remove non-alphanumeric characters,
        # time complexity: O(n) since each character is visited.
        # if non-alphanumeric character, then '': O(1)
        # if non-alphanumeric character, then leave as it is: O(1)
        cleaned_string = re.sub(r'[^a-z0-9]', '', lowercase_string)


        # The outer loop is O(m/2), where m is the number of alphanumeric characters, m <= n
        # This is O(m/2) since half of m characters is visited.
        for index in range(len(cleaned_string) // 2):
            if cleaned_string[index] != cleaned_string[len(cleaned_string) - index - 1]: # This is O(1), as constant operation
                return False

        # The total time complexity is: O(n) + O(n) + O(m/2) = O(n)

        return True
````

````python
# check with examples
s = "A man, a plan, a canal: Panama"
solution = Solution()

result = solution.isPalindrome(s)
print(result) # True
# Explanation: "amanaplanacanalpanama" is a palindrome.
````

````python
# check with examples
s = "race a car"
solution = Solution()

result = solution.isPalindrome(s)
print(result) # False
````

````python
 s = " "
 solution = Solution()

result = solution.isPalindrome(s)
print(result) # True
````
