Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

**Example 1:**

Input: nums = [2,7,11,15], target = 9

Output: [0,1]

Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

**Example 2:**

Input: nums = [3,2,4], target = 6

Output: [1,2]

**Example 3:**

Input: nums = [3,3], target = 6

Output: [0,1]

# Answer

# For LeetCode Submission:

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        # using two-nested for loops
        for index in range(len(nums)):
          for index2 in range(index + 1, len(nums)):
            if nums[index] + nums[index2] == target:
              return [index, index2]

        return None
```

```python
# Testing
nums = [2,7,11,15]
target = 9
# output: [0, 1]

# intialize
solution = Solution()

solution.twoSum(nums, target)
print(solution.twoSum(nums,target))
```

# Analysis

Time complexity: $O(n^2)$ because there are two nested loops.

# Code review
1. len(nums) calculates the number of elements in the array nums. Example: len([4,7,13]) = 3

2. range(y) would return the sequence of numbers [0, 1, 2, ... , y - 1]. Example: range(5) = [0, 1, 2, 3, 4]. 5 is not included.

3. in the second nested loop, range(index + 1, ...) has + 1 because it is convenient to visit the elements starting with the index + 1.

# Writing in class - for LeetCode submission

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        # using two-nested for loops
        for index in range(len(nums)):
          for index2 in range(index + 1, len(nums)):
            if nums[index] + nums[index2] == target:
              return [index, index2]

        return None
```

```python
# Testing
nums = [2,7,11,15]
target = 9
# output should be: [0, 1]

# intialize
solution = Solution()

solution.twoSum(nums, target)
print(solution.twoSum(nums,target))
```
