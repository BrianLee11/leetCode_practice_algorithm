Given an integer array nums of length n, you want to create an array ans of length 2n where ans[i] == nums[i] and ans[i + n] == nums[i] for 0 <= i < n (0-indexed).

Specifically, ans is the concatenation of two nums arrays.

Return the array ans.

**Example 1:**

- Input: nums = [1,2,1]
- Output: [1,2,1,1,2,1]
- Explanation: The array ans is formed as follows:
  - ans = [nums[0],nums[1],nums[2],nums[0],nums[1],nums[2]]
  - ans = [1,2,1,1,2,1]

**Example 2:**

- Input: nums = [1,3,2,1]
- Output: [1,3,2,1,1,3,2,1]
- Explanation: The array ans is formed as follows:
  - ans = [nums[0],nums[1],nums[2],nums[3],nums[0],nums[1],nums[2],nums[3]]
  - ans = [1,3,2,1,1,3,2,1]

#Submitted codes at LeetCode

```python
class Solution(object):
    def getConcatenation(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """

        # initialize a new array that holds concatenation of two nums arrays
        conc_nums = []

        for index in range(2*len(nums)):
            conc_nums.append(nums[index%len(nums)])

        return conc_nums
```

```python
# testing
solution = Solution()

nums = [1,2,3]
result = solution.getConcatenation(nums)
print(result)

nums = [1, 3, 2, 1]
result = solution.getConcatenation(nums)
print(result)
```

#Analysis: time complexity

1. **Initialization**. `conc_nums = [] `This initialization takes $O(1)$.

2. **Length Calculation**: `len(nums)` Counting the number of elements in the array takes $O(n)$, where n is the number of elements in nums.

3. **Loop for Concatenation:** `for index in range(2 * len(nums)):`
 Taking an index from (0, 2n) takes $O(2n)$, where n is the number of elements in nums.

4. **Appending Elements:** Inside the for loop, `the conc_nusm.append()` takes $O(1)$.

5. For total complexity of the for loop is $O(2n)$.

6. **The overall tiem complexity** The total time complexity is the sum of the individual steps:

 $$O(1) + O(2n) = O(2n) = O(n)$$

# Intuition / basic idea


Iterate and Append Elements:

Use a loop to iterate through the indices from 0 to 2n-1.

For each index i in this range:

- Compute the corresponding element in nums using the `index i % n`. This ensures that after reaching the end of the original nums array, the index wraps around to the beginning of nums.
- Append the element `nums[i % n]` to ans.
