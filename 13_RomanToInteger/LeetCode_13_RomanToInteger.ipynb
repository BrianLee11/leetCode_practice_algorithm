**Title: 13. Roman to integer**

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Each symbol respresents values in these orders:

- I = 1
- V = 5
- X = 10
- L = 50
- C = 100
- D = 500
- M = 1000

For example, 2 is written in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right.

However, the numeral for four is not IIII. Instad, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances, where subtraction is used:

- I can be placed before V (5) and X (10) to make 4 and 9.

- X can be placed before L (50) and C (100) to make 40 and 90.

- C can be placed before D (500) and M (1000) to make 400 and 900.

Therefore six exceptional cases would be:

1. IV: 5
2. IX: 9
3. XL: 40
4. XC: 90
5. CD: 500
6. CM: 900


**Task: given a roman numeral, convert it to an integer.**

#Answer

# Submission to LeetCode

```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """

        # dictionary to save corresponding score
        dictRomanToInt = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}

        # initialize score, string
        score = 0
        string = s

        def exception(string):
          if string[0:2] == 'IV' or string[0:2] == 'IX' or string[0:2] == 'XL' or string[0:2] == 'XC' or string[0:2] == 'CD' or string[0:2] == 'CM':
            return True
          else:
            return False

        while string != []:
          if len(string) == 1:
            score = score + dictRomanToInt[string[0]]
            string = []
          elif len(string) >= 2:
            if exception(string) == True:
                score = score - dictRomanToInt[string[0]] + dictRomanToInt[string[1]]
                if len(string) >= 3:
                  string = string[2:]
                else:
                  string = []
            else:
              score = score + dictRomanToInt[string[0]]
              string = string[1:]
        return score
```

```python
# testing with examples
solution = Solution()

print(solution.romanToInt('III'))     # output should be 3
print(solution.romanToInt('LVIII'))   # output should be 58
print(solution.romanToInt('MCMXCIV')) # output should be 1994
```
