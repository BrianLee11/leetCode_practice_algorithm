**Question 217: contains duplicate**

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

**Example 1:**

- Input: nums = [1,2,3,1]
- Output: true

**Example 2:**

- Input: nums = [1,2,3,4]
- Output: false

**Example 3:**
- Input: nums = [1,1,1,3,3,4,3,2,4,2]
- Output: true

#**Codes submitted to LeetCode**

````python
class Solution(object):
    def containsDuplicate(self, nums):

        # initialize a set
        # this is O(1) time complexity
        seen = set()

        # the total time
        for number in nums:     # O(n) as each element in nums is visited
            if number in seen:  # O(1) as search in set is O(1)
                return True
            else:
                seen.add(number) # O(1) as constant operation

        return False
````

````python
# Testing
nums = [1,2,3,1]
solution = Solution()

result = solution.containsDuplicate(nums)
print(result) # should be true
````

````python
# Testing
nums = [1,2,3,4]
solution = Solution()

result = solution.containsDuplicate(nums)
print(result) # should be false
````

````python
# Testing
nums = [1,1,1,3,3,4,3,2,4,2]
solution = Solution()

result = solution.containsDuplicate(nums)
print(result) # should be true
````

#**Intuition**

- the data structure set stores only unique values
- for each nubmer in nums, check whether any number after this number has the same value.
- If found, return True
- If not found, return False .
