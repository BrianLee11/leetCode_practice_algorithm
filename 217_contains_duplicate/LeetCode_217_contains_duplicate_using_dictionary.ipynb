**Question 217: contains duplicate**

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

**Example 1:**

- Input: nums = [1,2,3,1]
- Output: true

**Example 2:**

- Input: nums = [1,2,3,4]
- Output: false

**Example 3:**
- Input: nums = [1,1,1,3,3,4,3,2,4,2]
- Output: true

#**Codes submitted to LeetCode**

```python
class Solution(object):
    def containsDuplicate(self, nums):
        """
        input: an array of number
        output: bool
        time complexity: O(n), where n is the number of elements in nums
        intuition: in a hash-map (dictionary in Python), save an element as key and its frequency as value.
        Save the dictionary's values as list. Iterate over the values and check whether there is a value greater than 1.
        When this value is discovered, return True. Otherwise, return False.
        """

         # initialize a dictionary to store key-value
         # key would be a number in nums
         # value would be the frequency of key(number)
         # time complexity: O(1), as a constant operation
        dictionary = {}

        # construct key-value of the dictionary
        # if number exists in dictionary's keys, increment its value by 1
        # if number does not exist in keys, initialize its value by 1
        # time complexity: O(n), where n = nubmer of elements in the nums, since visiting every element in the nums
        for number in nums:
            if number in dictionary:
                dictionary[number] += 1
            else:
                dictionary[number] = 1

        # initialize a list to store values of dictionary
        # time complexity: O(m), where m is the unique elements of nums, since each value is converted as an element of list
        list_dictionary = list(dictionary.values())

        # discover whether frequency larger than 1 exist in the list
        # true whether value larger than 1 is found
        # false otherwise
        # time complexity: O(m)
        for value in list_dictionary:
            if value > 1:
                return True

        return False
```

```python
# Testing
nums = [1,2,3,1]
solution = Solution()

result = solution.containsDuplicate(nums)
print(result) # should be true
```

```python
# Testing
nums = [1,2,3,4]
solution = Solution()

result = solution.containsDuplicate(nums)
print(result) # should be false
```

```python
# Testing
nums = [1,1,1,3,3,4,3,2,4,2]
solution = Solution()

result = solution.containsDuplicate(nums)
print(result) # should be true
```

#**Intuition**

- In a hash-map (dictionary in Python), save an element as key and its frequency as value.
- Save the dictionary's values as list. Iterate over the values and check whether there is a value greater than 1.
- When this value is discovered, return True. Otherwise, return False.
